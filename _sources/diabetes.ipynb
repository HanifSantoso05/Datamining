```python
from scipy.io.arff import loadarff 
from io import StringIO, BytesIO
import urllib.request

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00329/messidor_features.arff'
resp = urllib.request.urlopen(url)
data, meta = loadarff(StringIO(resp.read().decode('utf-8')))
df = pd.DataFrame(data)

df
```

```python
col_names = []
for i in range(20):
    if i == 0:
        col_names.append('quality')
    if i == 1:
        col_names.append('prescreen')
    if i >= 2 and i <= 7:
        col_names.append('ma' + str(i))
    if i >= 8 and i <= 15:
        col_names.append('exudate' + str(i))
    if i == 16:
        col_names.append('euDist')
    if i == 17:
        col_names.append('diameter')
    if i == 18:
        col_names.append('amfm_class')
    if i == 19:
        col_names.append('label')
```

```python
df.columns = [col_names]
df
```

```python
X = df.drop(columns=['label'])
X
```

## Normalisasi
### Normalisasi Min and Max

$$ X'= \frac{x-x_{min}}{x_{max}-x_{min}}$$

```python
from sklearn.preprocessing import MinMaxScaler
import joblib

scaler = MinMaxScaler()
#scaler.fit(features)
#scaler.transform(features)
scaled = scaler.fit_transform(X)
features_names = X.columns.copy()
#features_names.remove('label')
scaled_features = pd.DataFrame(scaled, columns=features_names)
```

```python
scaled_features
```

### Save Model Preprocessing

```python
filename = "normalisasi.sav"
joblib.dump(scaler, filename) 
```

## Encoder Label

```python
y = df['label'].values
y[0:5]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
```

```python
yBaru = le.transform(y)
```

```python
yBaru
```

## Split Dataset

split datanya 80% 20% : 20% (data test), 80% (data training)

```python
from sklearn.model_selection import train_test_split
training, test = train_test_split(scaled_features,test_size=0.2, random_state=1)#Nilai X training dan Nilai X testing
training_label, test_label = train_test_split(yBaru, test_size=0.2, random_state=1)#Nilai Y training dan Nilai Y testing
```

## Gaussian Naive Bayes

```python
#Model Select
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(training, training_label)

probas = gaussian.predict_proba(test)
probas = probas[:,1]
probas.shape
```

```python
training.shape
```

```python
probas
```

```python
test_label
```

```python
probas = probas.round()

probas
```

```python
cm = confusion_matrix(test_label, probas)
cm
```

```python
print(f'acuracy = {accuracy_score(test_label, probas) * 100 :.2f}')
```

### Save Mode Gaussian


```python
filemodelGaussian = 'modelGaussian.pkl'
joblib.dump(gaussian, filemodelGaussian)
```

## K-Nearest Neighbour (KNN)

```python
#KNN Library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
my_param_grid = {'n_neighbors':[2,3,5,7], 'weights': ['distance', 'uniform']}
GridSearchCV(estimator=KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
knn = GridSearchCV(KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
```

```python
knn.fit(training, training_label)
```

### Save Model KNN

```python
filenameModelKnn = 'modelKnn.pkl'
joblib.dump(knn, filenameModelKnn)
```

```python
pred_test = knn.predict(test)
```

```python
print(f'acuracy = {accuracy_score(test_label, pred_test) * 100 :.2f}')
```
